The relation between the Total Population and Total Crimes Incidence are somewhat linear.
So, we can try to build the simple linear regression model on Total Crimes Incidence and Total Population where we will try to predict the population of the place with the help of Total no of Crime Incidents happened there

## **Data Preparation** ##

```python
x = df['Total Crimes Incidence']
y = df['Total Population']
```

```python
print(x.shape)
x.head()
```

```python
print(y.shape)
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## **Model Building** ##

```python
x_train_sm = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train_sm).fit()
```

```python
print(model.summary())
```

The linear equation for the model : Total Population  = 429.5801 * Total Crimes Incidence + 4611000

```python
plt.scatter(x_train,  y_train)
plt.plot(x_train, 4611000 + x_train * 429.5801, 'r')
plt.show()
```

## **Residual Analysis** ##

```python
y_train_pred = model.predict(x_train_sm)
```

```python
y_train_pred.head()
```

```python
residual = (y_train - y_train_pred)
residual.head()
```

```python
sns.distplot(residual)
```

## **Prediction on the test data** ##

```python
x_test_sm = sm.add_constant(x_test)
y_pred = model.predict(x_test_sm)
```

```python
y_pred.head()
```

## **Root mean squared error:** ##

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
RMSE
```

## **R2 score:** ##

```python
r2_score(y_test, y_pred)
```
